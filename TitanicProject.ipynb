```python
result = df[df['Survived']==1]['Pclass'].value_counts().sort_index()
print("The survival of passenger class\n",result)
```

```python
fare_surv = df[df['Survived']==1]['Fare'].mean()
fare_nons = df[df['Survived']==0]['Fare'].mean()
print("The mean fare of survived is: ",fare_surv)
print("The mean fare of non-survived is: ",fare_nons)
```

```python
sur_age = df[df['Survived']==1]['Age'].mean()
nonsur_age = df[df['Survived']==0]['Age'].mean()
print("No fifference in age of survival and non survivals ",sur_age,"\t",nonsur_age)
```
